```python
#Task 1 
#Sort the data as per total number of cases

import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt
dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
sorted_df = df.sort_values(by = ['total_cases'])
sorted_df
```

```python
#Task 2
#Get top 5 countries who has the least number of cases and plot a bar graph
lowest_cases = sorted_df['total_cases'].head(5)
lowest_country = sorted_df['country'].head(5)
print(lowest_cases)
print(lowest_country)
plt.ylabel("Country")
plt.xlabel("Number Of Cases")
plt.xticks(rotation = 'vertical')

plt.bar(lowest_country, lowest_cases, width = 0.4, color = ("red", "orange", "yellow", "green", "blue"))
```

```python
#Task 3
#Sort the data as per total number of deaths

sorted_deaths = df.sort_values(by = ['total_deaths'])
sorted_deaths
```

```python
#Task 4
#Get top 5 countries who has the maximum number of deaths and plot a bar graph

highest_deaths = sorted_deaths['total_deaths'].tail(5)
highest_deaths_country = sorted_deaths['country'].tail(5)
print(highest_deaths)
print(highest_deaths_country)
plt.ylabel("Country")
plt.xlabel("Number Of Deaths")
plt.xticks(rotation = 'vertical')

plt.bar(highest_deaths_country, highest_deaths, width = 0.4, color = ("red", "orange", "yellow", "green", "blue"))
```

```python
#Task 5
#Sort the data as per active cases
sorted_cases = df.sort_values(by = ['active_cases'])
sorted_cases
```

```python
#Task 6
#Get top 5 countries who has the maximum number of active cases and plot a bar graph
highest_active_cases = sorted_cases['active_cases'].tail(5)
highest_active_cases_country = sorted_cases['country'].tail(5)
print(highest_active_cases)
print(highest_active_cases_country)
plt.ylabel("Country")
plt.xlabel("Number Of Active Cases")
plt.xticks(rotation = 'vertical')

plt.bar( highest_active_cases_country, highest_active_cases, width = 0.4, color = ("red", "orange", "yellow", "green", "blue"))
```
